```python
#initialize variables
tks = {}
```

```python
#parameters
exp_time = 0.1 #exposure time in s (was 100ms)
px_size = 65 #pixel size in nm
'''
set parameters here e.g.
min_vel =
max_vel = 
min_snr =
min_tk_len =
max_tk_len = 

'''
```

```python
start_time = time.time()

while True:
    line = infile.readline()
    
    if line:
        if 'ID' in str(line.split(',')[0]):
            spot_id = int(line.split(',')[1])  #spot ID
            if len(line.split(',')[2]) == 0:
                tk_id = -1 #for now I assign all spots without a track ID to have a track ID of -1
            else:
                tk_id = int(line.split(',')[2])  #track ID
            x = float(line.split(',')[4]) #x position in units specified (e.g. if image parameters set to nm, it will use that)
            y = float(line.split(',')[5]) #y position in units specified (e.g. if image parameters set to nm, it will use that)
            t = float(line.split(',')[8]) #time position in units frames
            mean_intens = float(line.split(',')[12])
            tot_intens = float(line.split(',')[16])
            snr = float(line.split(',')[19])
            xy = (x,y)

            if tk_id in tks.keys():
            # append the new number to the existing array
                tks[tk_id].append([spot_id,xy])
            else:
            # create a new array in this slot
                tks[tk_id] = [[spot_id,xy]]
            
    if not line:
        break
infile.close()

print ('Number of tracks is: ' + str(len(tks)))

print('Tracks read in %s seconds' % (time.time() - start_time))
```
